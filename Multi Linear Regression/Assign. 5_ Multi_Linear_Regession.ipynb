```python
# import nessesary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api  as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
# Model validation
# 1)collinearity using vif, vif=(1/(1-R^2)) for all independent variables
rsqr_a=smf.ols("ADMS~RDS+MKTS",data=df1).fit().rsquared
vif_a=1/(1-rsqr_a)

rsqr_m=smf.ols("MKTS~RDS+ADMS",data=df1).fit().rsquared
vif_m=1/(1-rsqr_m)


rsqr_r=smf.ols("RDS~MKTS+ADMS",data=df1).fit().rsquared
vif_r=1/(1-rsqr_r)

d1={'Variables':['ADMS','MKTS','RDS'],'VIF':[vif_a,vif_m,vif_r]}
Vif_df=pd.DataFrame(d1)

Vif_df

```

```python
#2)Residual Analysis
#Test for normality of residuals(Q-Q plot) using mode..resid
sm.qqplot(model.resid, line='q')
plt.title("Normal Q-Q Plot of Residuals")
plt.show()
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show() 
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```

```python
# 2) Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k=df1.shape[1]
n=df1.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
df1[df1.index.isin([49])]
```

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
df2=df1.drop(df1.index[[49]],axis=0).reset_index(drop=True)
df2
```

```python
# Model Deletion Diagnostics and Final Model
model2=smf.ols("Profit~RDS+ADMS+MKTS",data=df2).fit()
```

```python
while model2.rsquared < 0.99:
    for c in [np.max(c)>1]:
        model2=smf.ols("Profit~RDS+ADMS+MKTS",data=df2).fit()
        (c,_)=model2.get_influence().cooks_distance
        c
        np.argmax(c) , np.max(c)
        df2=df2.drop(df2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        df2
    else:
        final_model=smf.ols("Profit~RDS+ADMS+MKTS",data=df2).fit()
        final_model.rsquared , final_model.aic
        print("Thus model accuracy is improved to",final_model.rsquared)
```
